```python
aux.groupby(['Accession']).count()
```

```python
aux = accesssions[accesssions['Accession'].isin(seq3['accession_number'].values)]
for a in set(aux['Accession']):
    seq3.loc[seq3['accession_number']==a,"E Ct"] = aux.loc[aux['Accession']==a,"E Cts"].values[0]
    seq3.loc[seq3['accession_number']==a,"N1 Ct"] = aux.loc[aux['Accession']==a,"N Cts"].values[0]
    seq3.loc[seq3['accession_number']==a,"N2 Ct"] = aux.loc[aux['Accession']==a,"N Cts"].values[0]
    seq3.loc[seq3['accession_number']==a,"RdRp Ct"] = aux.loc[aux['Accession']==a,"RdRp Cts"].values[0]
    seq3.loc[seq3['accession_number']==a,"RNAse P Ct_internal"] = aux.loc[aux['Accession']==a,"RNAse P Cts"].values[0]
```

```python
seq3.to_csv('../../../../Downloads/COMET384-L-003-with-cts.csv')
```
